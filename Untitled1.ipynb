```python
s = input( "enter 3 real number ( separated by space ) : ")
s = s.split ()
a = float(s[0])
b = float(s[1])
c = float(s[2])
print(f"a ={a} , b = {b} , c = {c}")
if a == 0:
    if b == 0:
        print("Phương trình không hợp lệ.")
    else:
        x = -c / b
        print("Phương trình bậc nhất có nghiệm:")
        print(f"x = {x}")
else:
    delta = b**2 - 4*a*c
    print(f"Delta = {delta}")

    if delta > 0:
        x1 = (-b + delta**0.5) / (2*a)
        x2 = (-b - delta**0.5) / (2*a)
        print("Phương trình có hai nghiệm phân biệt:")
        print(f"x1 = {x1}")
        print(f"x2 = {x2}")
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép:")
        print(f"x = {x}")
    else:
        print("Phương trình vô nghiệm trong tập số thực.")

```
